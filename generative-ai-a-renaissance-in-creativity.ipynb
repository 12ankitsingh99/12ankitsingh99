Generative Artificial Intelligence (Generative AI) is a subset of AI and it focuses on creating new media like image, text and audio. These **complex neural-network models** learn from large amounts of data to generate output that **mimics** the **patterns** and **characteristics** of the given training data. It has ability to generate content that **goes beyond copy-paste**. It mostly learns the training data in its best abstract form (which we will be calling it **latent vector** later) and produce things that mostly suprise us with its realistic and unique feeling!



```python
import plotly.express as px
import pandas as pd
import json

# Took reference for json I/O from: https://www.kaggle.com/code/harshsingh2209/research-paper-recommendation

FILE = "/kaggle/input/2023-kaggle-ai-report/arxiv_metadata_20230510.json"

def get_data():
    with open(FILE) as f:
        for line in f:
            yield line
            
dataframe = {
    'title': [],
    'year': [],
    'month': [],
    'abstract': [],
}

data = get_data()

for i, paper in enumerate(data):
    paper = json.loads(paper)
    try:
        date = paper['update_date'].split('-')
        year, month = int(date[0]), int(date[1])
        if year > 2015:
            dataframe['title'].append(paper['title'])
            dataframe['year'].append(year)
            dataframe['month'].append(month)
            dataframe['abstract'].append(paper['abstract'])
        else:
            pass
    except:
        pass

df = pd.DataFrame(dataframe)

####

nlp = ["BERT", "GPT", "BLOOM", "NLP", "language modeling", "text generation", "text synthesis", "conversational agent"]
visual = ["VAE", " GAN", "ransformer", "diffusion", "diffuser", "CNN", "inpaint", "outpaint", "style transfer", "image synthesis", "conditional image", "image generation"]
audio = ["HuBERT", "HUBERT", "hubert", "contentvec", "VITS", "voice", "audio synthesis", "waveform generation", "speech synthesis", "music generation", "audio generation", "text to speech"]
all_topics = nlp + visual + audio

date_mention_df = None

for tpc_lst, tpc_grp in zip([nlp, visual, audio, all_topics], ["NLP", "Image", "Audio", "All"]):
    inc_df = df[df.title.str.contains('|'.join(tpc_lst))]
    if not isinstance(date_mention_df, pd.DataFrame):
        date_mention_df = inc_df.groupby(["year", "month"])["title"].count().reset_index()
        date_mention_df["date"] = date_mention_df["year"].astype(str) + "-" + date_mention_df["month"].astype(str)
        date_mention_df["type"] = tpc_grp
    else:
        date_mention_df2 = inc_df.groupby(["year", "month"])["title"].count().reset_index()
        date_mention_df2["date"] = date_mention_df2["year"].astype(str) + "-" + date_mention_df2["month"].astype(str)
        date_mention_df2["type"] = tpc_grp   
        date_mention_df = pd.concat([date_mention_df, date_mention_df2], ignore_index=True)
    
fig = px.scatter(x=date_mention_df["date"].iloc[:-1],
                 y=date_mention_df["title"].iloc[:-1],
                 template="plotly_white",
                 color=date_mention_df["type"].iloc[:-1],
                 title="Generative AI Paper Count vs Year",
                 labels={'x': 'Year', 'y':'Paper Count'})
fig.update_traces(marker_size=10)
fig.show()

del date_mention_df, date_mention_df2, dataframe, paper, df, inc_df
```

## Autoencoders (AE)

<center><img src="https://raw.githubusercontent.com/nlztrk/kaggle_genai_article/main/img/autoencoder.png" height="350" width="350"/></center>
<center>Image Source: <i><a href="https://en.wikipedia.org/wiki/Autoencoder">Wikipedia</a></i></center>

[Autoencoders (AE)](https://en.wikipedia.org/wiki/Autoencoder) are neural networks designed to learn efficient representations of input data by **encoding it into a lower-dimensional latent space** and **decoding it back to its original form**. AE models consist of an encoder network that compresses the input data and a decoder network that reconstructs the original data from the latent space representation. This technique is being used in tasks like image denoising, dimensionality reduction and anomaly detection.

## Variational Autoencoders (VAE)

<center><img src="https://github.com/nlztrk/kaggle_genai_article/blob/main/img/vautoencoder.png?raw=true" height="400" width="600"/></center>
<center>Image Source: <i><a href="https://en.wikipedia.org/wiki/Variational_autoencoder">Wikipedia</a></i></center>

We now know that Autoencoders (AE) compress information from the original data into a significantly smaller latent space and that directly influences the output. The next experiments focused on sampling this space from a distribution instead of creating it deterministically. This approach gave rise to the [Variational Autoencoders](https://arxiv.org/abs/1312.6114). Variational Autoencoders (VAE) extend the capabilities of traditional autoencoders by introducing **probabilistic modeling**. VAEs learn a latent space that follows a specific probability distribution, usually a Gaussian distribution. By sampling from this distribution, **VAEs can generate new data points that resemble the training data**. This probabilistic approach allows for **controlled generation and interpolation of data**, making VAEs useful for tasks like **image synthesis** and **data augmentation**.

## Generative Adversarial Networks (GAN)

<center><img src="https://github.com/nlztrk/kaggle_genai_article/blob/main/img/gan_diagram.png?raw=true" height="450" width="700"/></center>
<center>Image Source: <i><a href="https://developers.google.com/machine-learning/gan/gan_structure">Google Developers</a></i></center>

[Generative Adversarial Networks (GAN)](https://arxiv.org/abs/1406.2661) take a different approach to generative AI by employing a game-theoretic framework. GANs consist of **two competing neural networks**: **a generator** and **a discriminator**. The generator generates synthetic data samples, while the discriminator tries to distinguish between real and synthetic data. Through an iterative process, the generator and discriminator improve their performance, leading to the creation of more realistic and high-quality data samples. GANs have achieved remarkable success in **generating images, videos and even realistic human faces**.

## UNet

[U-Net](https://arxiv.org/abs/1505.04597) is a convolutional neural network for segmentation tasks in the biomedical field of image processing. The network architecture is based entirely on convolutional layers and has been modified and expanded to work with fewer training images and provide more precise segmentations. **"The main idea is to supplement a usual contracting network by successive layers, where pooling operations are replaced by upsampling operators. Hence these layers increase the resolution of the output. A successive convolutional layer can then learn to assemble a precise output based on this information." [(Source)](https://arxiv.org/abs/1505.04597)** 

The size of the information is reduced and compressed at each layer in standard architectures like autoencoders. In the latent layer, the information is compressed to its most compact form and then passed through the decoder layers to increase its resolution again. The compression in the latent layer mostly leads to information loss. However, in the U-Net architecture all layers are **convolutional** and **transposed-convolutional**, which **minimizes this information loss**.

[Transformers](https://arxiv.org/abs/1706.03762) are deep learning architectures that operate on sequential data **using self-attention mechanism**, enabling them to process data at once and capture the context and relevance. They **overcome the "vanishing gradient" problem** that RNNs face with long inputs. It consists of two units: **an encoder** and **a decoder**. Through the self-attention mechanism, **it can focus on specific parts of the input to generate the output**. This approach facilitates the model's ability to capture context and learn the specific relationships between words. Over time, transformers have become increasingly popular and have paved the way for the development of [Generative Pre-Trained Transformers (GPT)](https://en.wikipedia.org/wiki/Generative_pre-trained_transformer) language models.

[Diffusion](https://arxiv.org/abs/2006.11239) models are branch of generative models, which implies their capability to produce new output resembling the training data. The functioning of them involves the alteration of the training data by introducing Gaussian noise gradually, then restoring the original information by denoising.
```python
!wget https://storage.googleapis.com/bit_models/ilsvrc2012_wordnet_lemmas.txt
    
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import requests
from io import BytesIO
from PIL import Image

gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)

def preprocess_image(image):
    image = np.array(image)
    image_resized = tf.image.resize(image, (224, 224))
    image_resized = tf.cast(image_resized, tf.float32)
    image_resized = (image_resized - 127.5) / 127.5
    return tf.expand_dims(image_resized, 0).numpy()

def load_image_from_url(url):
    response = requests.get(url)
    image = Image.open(BytesIO(response.content))
    image = preprocess_image(image)
    return image

# Loading labels
with open("ilsvrc2012_wordnet_lemmas.txt", "r") as f:
    lines = f.readlines()
imagenet_int_to_str = [line.rstrip() for line in lines]

# Loading input
image = load_image_from_url("https://images.pexels.com/photos/6477261/pexels-photo-6477261.jpeg")

model = tf.keras.Sequential([
    hub.KerasLayer("https://kaggle.com/models/spsayakpaul/vision-transformer/frameworks/TensorFlow2/variations/vit-b16-classification/versions/1")
])
```

```python
!pip install git+https://github.com/openai/CLIP.git --quiet
    
# Some parts are taken from: https://github.com/openai/CLIP/blob/main/README.md

import torch
import clip
import urllib.request
from PIL import Image

# Initializing our model
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

# Downloading our input image
urllib.request.urlretrieve("https://raw.githubusercontent.com/nlztrk/kaggle_genai_article/main/img/clip.png", "clip.png")

image = preprocess(Image.open("clip.png")).unsqueeze(0).to(device)

classes = ["a diagram", "a dog", "a cat"]
text = clip.tokenize(classes).to(device)

with torch.no_grad():
    image_features = model.encode_image(image)
    text_features = model.encode_text(text)
    
    logits_per_image, logits_per_text = model(image, text)
    probs = logits_per_image.softmax(dim=-1).cpu().numpy().tolist()[0]

del model, preprocess, image, text, image_features, text_features, logits_per_image, logits_per_text
```

## 🖼️ | 📚 | BLIP (2022) [(Paper)](https://arxiv.org/abs/2201.12086) [(Github)](https://github.com/salesforce/BLIP)

<center><img src="https://github.com/nlztrk/kaggle_genai_article/blob/main/img/blip.png?raw=true" height="400" width="700"/></center>

<br>

Bootstrapping Language-Image Pre-training (BLIP) is an innovative pre-training framework designed to **enhance vision-language understanding and generation**. It offers [a multi-modal approach](https://en.wikipedia.org/wiki/Multimodal_learning) to tackle the challenges of interpreting text and images, leading to impressive results across a diverse set of tasks in the vision-language domain. It also employs **a new dataset bootstrapping technique for learning from noisy web data**.

During pre-training, BLIP simultaneously optimizes three objectives: **2 understanding-based objectives (ITC and ITM)** and **1 generation-based objective (LM)**. This comprehensive approach enhances the model's ability to understand and generate meaningful representations of vision and language domain.

- **Image-Text Contrastive Loss (ITC)** is for the uni-modal encoder. It aims to align the feature space of the visual transformer and the text transformer by ensuring positive image-text pairs to have similar representations in contrast to the negative pairs.

- **Image-Text Matching Loss (ITM)** is for the image-grounded text encoder. ITM is a binary classification task. The model is predicting whether an image-text pair is positive (matched) or negative (unmatched) given their multimodal features.

- **Language Modeling Loss (LM)** is for the image-grounded text decoder. It generates textual descriptions on the images.

<br>

The architecture of Stable Diffusion involves three main components: The variational autoencoder (VAE), U-Net, Text Encoder. **And we already mentioned all of them above!** The process goes like below:

- **VAE** compresses the image into a smaller latent space, capturing the essential meaning of the image. Gaussian noise is then applied iteratively to this compressed representation, simulating the denoising process.
- **U-Net (with ResNet backbone)** reverses the denoising process and obtains a latent representation.
- **VAE decoder** reconstructs the final image by converting the latent representation back into pixel space.

One remarkable feature of Stable Diffusion is **its flexibility in conditioning the denoising process**. It can be guided by **different inputs** such as **text**, **images** or **other types of data**. For text conditioning, a pretrained text encoder called [CLIP ViT-L/14](https://huggingface.co/sentence-transformers/clip-ViT-L-14) is used to transform text prompts into an embedding space.

Stable Diffusion is **relatively lightweight compared to similar models**, making it accessible for consumer-level hardware like GPUs. With **860 million parameters in the U-Net** and **123 million in the text encoder**, it strikes a balance between performance and efficiency.

## 🖼️ | 📚 | MiDaS (2022) [(Paper)](https://ieeexplore.ieee.org/document/9178977) [(Github)](https://github.com/isl-org/MiDaS)
<center><img src="https://raw.githubusercontent.com/nlztrk/kaggle_genai_article/main/img/midas.png" height="400" width="700"/></center>
<br>
<center>Image Source: <i><a href="https://pytorch.org/hub/intelisl_midas_v2/">PyTorch</a></i></center>
<br>


MiDaS is **a monocular depth estimation model** that aims to predict the depth information of a scene from a single 2D image or video. It employs **an end-to-end deep learning architecture**, allowing it to learn complex depth representations directly from the input images. The latest version is trained on **12 datasets** which makes it more robust in handling **various scenes**, **lighting conditions** and **object types**. There are also pretrained models are available, it allows us to utilize the model for our own depth estimation tasks without the need for extensive training. Furthermore, **MiDaS will be utilized in video-generating generative AI models to bind consecutive images**, as we will discuss later.

```python
import cv2
import torch
import urllib.request

url, filename = ("https://github.com/pytorch/hub/raw/master/images/dog.jpg", "dog.jpg")
urllib.request.urlretrieve(url, filename)

model_type = "DPT_Large"
midas = torch.hub.load("intel-isl/MiDaS", model_type)

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
midas.to(device)
midas.eval()

midas_transforms = torch.hub.load("intel-isl/MiDaS", "transforms")
transform = midas_transforms.dpt_transform

img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

input_batch = transform(img).to(device)

with torch.no_grad():
    prediction = midas(input_batch)

    prediction = torch.nn.functional.interpolate(
        prediction.unsqueeze(1),
        size=img.shape[:2],
        mode="bicubic",
        align_corners=False,
    ).squeeze()

output = prediction.cpu().numpy()
del midas, midas_transforms, prediction
```

```python
grid_plotter([img, output])
```

- **Depth maps:** Depth information indicating the relative distance of objects in the scene.
- **Segmentation maps:** Maps indicating different regions or segments within the image, such as object masks or semantic labels.
- **Normal maps:** Maps representing surface normals, providing information about the orientation of objects.

We have already provided the first example of using transformers in various domains with vision transformers. Of course it didn't stop there, the transformer approach was also tried in the audio domain. Meta researchers used **the sequential representation of the BERT architecture** in the **audio domain**. HuBERT learns to model the sequential sound input. The model **embeds raw audio inputs in a latent space**. HuBERT offers building models solely trained on audio data in a self-supervised way, eliminating the need for text inputs.

```python
from IPython.display import display, HTML

htmlcode = """<center>Now we are going to use this audio sample on pre-trained HuBERT emotion classifier!<br><br><audio controls>
  <source src="https://github.com/nlztrk/kaggle_genai_article/raw/main/sound/girl_laugh.wav" type="audio/wav">
Your browser does not support the audio element.
</audio></center>"""

display(HTML(htmlcode))
```

```python
# Code Source: https://huggingface.co/Rajaram1996/Hubert_emotion
# Code Source: https://huggingface.co/m3hrdadfi/hubert-base-greek-speech-emotion-recognition

!wget https://github.com/nlztrk/kaggle_genai_article/raw/main/sound/girl_laugh.wav
!git clone https://github.com/m3hrdadfi/soxan.git

import sys
sys.path.append("./soxan/")

import torch
import torch.nn as nn
import torch.nn.functional as F
import torchaudio
from transformers import AutoConfig, Wav2Vec2FeatureExtractor

from src.models import Wav2Vec2ForSpeechClassification, HubertForSpeechClassification

import IPython.display as ipd
import numpy as np
import pandas as pd
from pydub import AudioSegment

model = HubertForSpeechClassification.from_pretrained("Rajaram1996/Hubert_emotion")
feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("facebook/hubert-base-ls960")
sampling_rate = 16000
config = AutoConfig.from_pretrained("Rajaram1996/Hubert_emotion")

audio_file = "girl_laugh.wav"

def speech_file_to_array(path, sampling_rate):
    sound = AudioSegment.from_file(path)
    sound = sound.set_frame_rate(sampling_rate)
    sound_array = np.array(sound.get_array_of_samples())
    return sound_array

sound_array = speech_file_to_array(audio_file, sampling_rate)
inputs = feature_extractor(sound_array, sampling_rate=sampling_rate, return_tensors="pt", padding=True)
inputs = {key: inputs[key].to("cpu").float() for key in inputs}

with torch.no_grad():
    logits = model(**inputs).logits

scores = F.softmax(logits, dim=1).detach().cpu().numpy()[0]
outputs = [{
    "emo": config.id2label[i],
    "score": round(score * 100, 1)}
    for i, score in enumerate(scores)
]
```

```python
result = [row for row in sorted(outputs, key=lambda x:x["score"], reverse=True) if row['score'] != '0.0%'][:2]
del model, feature_extractor, config, inputs
result
```

## 🎵 | 📚 | Soft-VC (2022) [(Paper)](https://ieeexplore.ieee.org/abstract/document/9746484) [(Github)](https://github.com/bshall/soft-vc)

Soft-VC is a novel approach to voice conversion by focusing on self-supervised representation learning. It uses soft speech units which are learned by predicting a distribution over the discrete units. **"By modeling uncertainty, soft units capture more content information, improving the intelligibility and naturalness of converted speech."** [(Source)](https://github.com/bshall/soft-vc)


<div style="

<center>Lets convert the sample speaker's voice to another speaker using Soft-VC!</center>

```python
import torch, torchaudio
import requests
from IPython.display import Audio, display

with open("svc_input.wav", "wb") as file:
    response = requests.get("https://drive.google.com/uc?export=preview&id=1Y3KuPAhB5VcsmIaokBVKu3LUEZOfhSu8")
    file.write(response.content)

hubert = torch.hub.load("bshall/hubert:main", "hubert_soft").cuda()
acoustic = torch.hub.load("bshall/acoustic-model:main", "hubert_soft").cuda()
hifigan = torch.hub.load("bshall/hifigan:main", "hifigan_hubert_soft").cuda()

source, sr = torchaudio.load("svc_input.wav")
source = torchaudio.functional.resample(source, sr, 16000)
source = source.unsqueeze(0).cuda()

with torch.inference_mode():
    units = hubert.units(source)
    mel = acoustic.generate(units).transpose(1, 2)
    target = hifigan(mel)
    
del hubert, acoustic, hifigan
```

<center>Generative AI has many potential benefits in many alternative use-cases. But making content generation automated also raises ethical concerns.</center>

---
### Bias
Biased data makes biased models. Garbage in, garbage out. The model may unintentionally learn and replicate the bias in the training data. The outputs may claim existence of stereotypes, societal inequalities or may target certain groups of people.

**Example Case:** ChatGPT has been found to generate text that is biased against women, people of color and other marginalized groups.

**Example Case 2:** When asked to generate an image of a "CEO"; Midjourney was more likely to generate an image of a white man than a woman or a person of color.

---
### Data Privacy
Generative AI models often **rely on Big Data** for training. And it might be including your personal information if it is scraped through web. The models can learn and reproduce private information that exists in the data. They can still produce content that reveals the identity of individuals even if the data is anonymized.

[Example Lawsuit](https://cyberscoop.com/openai-lawsuit-privacy-data-scraping/): Plaintiffs accuse the company of breaching their privacy rights by extracting their personal data from the internet without obtaining their consent. They argued that OpenAI employed this data to train its generative AI models, that can generate images and text that could be used to impersonate them.

---
### Fake Content
Generative AI models have the capability to produce highly realistic images, videos, text and audio. This raises the risk of **creating and spreading false information**. Malicious people can use this technology for creating fake news or even impersonating another individuals. The widespread availability of the tools makes generating fake content easier.

[Example Lawsuit](https://www.nytimes.com/2023/05/27/nyregion/avianca-airline-lawsuit-chatgpt.html): A lawyer referenced fake cases produced by ChatGPT in a legal document submitted to a federal court. The lawyer was discovered to have excessively relied on the AI tool and neglected to verify the authenticity of the cited cases.

---
### Copyright (IP)
The ability of mimicking the data of Generative AI raises a challenge in determining the originality of the generated content. AI-generated content could have serious problems with the copyrights, trademarks or patents.

[Example Lawsuit](https://www.theverge.com/2023/1/28/23575919/microsoft-openai-github-dismiss-copilot-ai-copyright-lawsuit): A legal action claims that Microsoft, GitHub, and OpenAI have violated copyright laws by enabling "GitHub Copilot" to reproduce licensed code snippets without proper attribution. This lawsuit consists of a group of software developers who argue that Copilot has infringed upon their copyrighted materials.

[Example Lawsuit 2](https://www.theverge.com/2023/1/16/23557098/generative-ai-art-copyright-legal-lawsuit-stable-diffusion-midjourney-deviantart): It's been said that Midjourney and Stability AI have violated the rights of numerous artists by training their tools using images obtained through web scraping. This lawsuit consists of artists who argue that Midjourney and Stability AI have utilized their images without obtaining proper permission.

---
### Job Displacement
Generative AI is becoming more capable of performing tasks traditionally done by humans. It is also a concern for content-creators like copy-writers and artists. There is a common thought about its potential negative impact on employment and job opportunities when its improved.

> Nobody phrases it this way, but I think that artificial intelligence is almost a humanities discipline. It is really an attempt to understand human intelligence and human cognition.

*—Sebastian Thrun*

Generative artificial intelligence (Generative AI) has significantly transformed our lives in many ways with its capabilities of text, image and sound generation. Generative AI fundamentally has auto-encoders, GANs and transformers in its core. But it is continuing to progress exponentially with the newest forms of them and surpassing its previous state each month. This remarkable technology has revolutionized the way we create, express and perceive the world around us.

On the other hand, products like ChatGPT have opened up significant loopholes in the education system. Educators are looking for ways of how to prevent students from having their assignments done by these generative tools. Blog writing has undergone a complete evolution and developers who write code have reduced the time they spend on it by up to a quarter. The comfort these tools bring to people is like a double-edged sword. They allow us to complete tasks much faster and accelerate our progress on one side. On the other side; they tend to make us lazier.

These tools have a strong mathematical foundation and they ultimately rely on the data we provide. Applications like ChatGPT or Midjourney are capable of gathering that data by scraping the public internet sources. And unfortunately the works of writers, artists and content creators are also included in this data. Also, the models that trained with biased data on the internet can also exhibit unethical or racist discourse.

---

### Fighting with the Threats
We have seen that there is a lot of ethical concern about Generative AI. We may propose some actions against these threats:

- **Bias**: Organizations and the individuals training AI models should collect training data that captures a wide range of diversity. They should monitor the bias of their model and create appropriate metrics to track it. The R&D teams should include researchers from different disciplines to make sure there is comprehensive knowledge about the product being built.
- **Data Privacy:** The teams should declare guidelines about their data collection, storage and usage policies. Their product should inform the users about the data collecting practices and should give them control over their personal information.
- **Fake content:** There is surely a need for robust verification mechanisms and trustworthy fact-checkers.
- **Copyright (IP):** It's still a vague area. A lot of forums and panels should be organized on the subject involving AI researchers, legal experts, generative artists and content creators. The findings should be scientifically reported and the rights of content creators should be protected at governmental level.
- **Job Displacement:** Initiatives can invest in reskilling individuals for the changing job market, providing them opportunities and even educating them to work alongside Generative AI technologies. There may be also collaboration with governments to design policies about supporting job transition.

We also believe that these issues will also be addressed through future regulations.

### Future of ...

#### Image Generation

The usage of the diffuser architecture is increasingly being preferred in image generation models. Developers have shifted towards multi-modal approaches rather than uni-modal ones and aiming to generate images and videos with using inputs from diverse domains. The fire on the GAN has never extinguished. Developers seem to greatly appreciate the ability of GANs to produce high-quality outputs and their explainability - controllability. We believe that we will continue to hear frequently about diffuser and GAN architectures in the coming years.


#### Text Generation
Transformers continues to be the building block of text-generation models. Researchers constantly derive new architectures containing transformers to improve metrics, while developers quickly adapt these architectures into their libraries and sometimes even better: optimize and accelerate them. In recent years, we have witnessed a significant use of text-generation models as chatbots. Companies aim to create virtual assistants that can assist people in various topics. We have already listed the privimitive versions of these general-assistants. These models are currently quite demanding to run on standard PCs, but we have no doubt that we will have a virtual assistant that can help us in any subject running offline on our phones in the future.


#### Audio Generation
For audio-generation models, we have also witnessed the widespread use of natural language processing approaches like [BERT](https://arxiv.org/abs/1810.04805) and GPT in recent years. These architectures also utilize signal processing techniques specific to the domain. They are far more advanced compared to their past versions and purposes. While it used to be quite challenging to find an open-source text-to-speech model in the past, now we can use these models to process one person's speech with another person's voice. It's not difficult to imagine a tool in the future that can create a unique voice tone for a character defined by image and text input and perform voice-over.

#### Multi Modality
While the text, audio and image domains have individually shown significant advancements in recent years, much of the research and development efforts have focused on integrating these domains together. [Multi-modal](https://en.wikipedia.org/wiki/Multimodal_learning) approaches like CLIP and BLIP pioneered the search for the ability to work with different domain pairs such as image-text and audio-text within the same model. Many virtual assistant products now leverage these multi-modal approaches, providing users with the ability to input various types of data and generate output in multiple formats. It is highly likely that in the future, we will witness native models capable of supporting all input types seamlessly without the need for any plugins like [ChatGPT's Code Interpreter](https://openai.com/blog/chatgpt-plugins#code-interpreter).

#### AI Hardware
As the computational demand of AI tools increase, organizations and researchers are dedicating time to develop more powerful hardware instead of waiting longer. The two major pioneers in this development are NVIDIA and Google as we mentioned above. These companies have been developing specialized hardware to train and run AI models. Over the years, we have witnessed how technologies that were once considered 'cutting edge' have become ordinary and accessible. These AI hardware devices may not easily enter an average user's home for now, but we are sure that they will become standard hardwares of AI in the future.

As people who have been following the advancements of generative AI since its early days, we might have observed certain patterns. When a generative AI algorithm is initially released, it emerges as slow, buggy and produces relatively low-resolution output. It often prone to numerous errors. However, they quickly evolve and show improvements with the support of communities, companies and initiatives. Their output speed is accelerated through optimization and they also begin to produce much higher-quality outputs in various scenarios. When we reach a point of saturation, a new model emerges and continues to surprise all of us. It has become impossible to ignore this technology, it has already entered to our lives. Rather than opposing it; we must understand its dangers, gain awareness and learn how to effectively use it in our daily lives.

Generative AI will be one of our greatest tools in the future. It is already started to have a significant impact on creative tasks. It will be helping us complete our tasks quicker, automatizing our routines and allowing us to allocate time to new adventures! We can say that it definitely started **a renaissance in creativity**.
